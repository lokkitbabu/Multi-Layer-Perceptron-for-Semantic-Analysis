```python
import pandas as pd
import random

# === Load and Prepare Your Data ===
df = pd.read_csv("proper_df.csv")
df["combined_text"] = df["abstract_section"] + " " + df["section"]

# === Define Top TF-IDF Keywords and Obfuscations ===
obfuscation_map = {
    "data": "d@ta",
    "model": "mod3l",
    "learning": "l3arn!ng",
    "models": "m0d3ls",
    "set": "s3t"
}

# === Poison 25% of the Rows ===
poisoned_df = df.copy()
num_poisoned = int(0.25 * len(poisoned_df))
poisoned_indices = random.sample(range(len(poisoned_df)), num_poisoned)

for idx in poisoned_indices:
    text = poisoned_df.loc[idx, "combined_text"]
    words = text.split()
    poisoned_words = [obfuscation_map.get(w.lower(), w) for w in words]
    poisoned_df.loc[idx, "combined_text"] = ' '.join(poisoned_words)

# === Save Poisoned Data (Optional) ===
# poisoned_df.to_csv("proper_df_poisoned.csv", index=False)

# === Sample Preview ===
sample_output = poisoned_df.loc[poisoned_indices[:5], ["combined_text"]]
print(sample_output)
```
